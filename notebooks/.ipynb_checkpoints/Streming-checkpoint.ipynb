```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```python
from pyspark.sql import SparkSession
import os

# 1. On s'assure de fermer une éventuelle session fantôme
try:
    if 'spark' in globals():
        spark.stop()
except:
    pass

# 2. Création de la nouvelle session
spark = (
    SparkSession
    .builder
    .master("local[*]")
    .appName("Stream From Kafka")
    .config("spark.cassandra.connection.host", "cassandra1,cassandra2,cassandra3") 
    .config("spark.cassandra.connection.port", "9042")
    .config("spark.streaming.stopGracefullyOnShutdown", "true")
    # Tous les packages sur une seule ligne
    .config("spark.jars.packages", 
            "org.apache.spark:spark-sql-kafka-0-10_2.12:3.5.1,"
            "com.datastax.spark:spark-cassandra-connector_2.12:3.5.0")
    .getOrCreate()
)

print("Spark Session créée avec succès !")
spark
```

```python
kafka_df = (
    spark
    .readStream
    .format("kafka")
    .option("kafka.bootstrap.servers","ed-kafka-1:29092,ed-kafka-2:29093,ed-kafka-3:29094")
    .option("subscribe","binance")
    .option("startingOffsets","earliest")
    .load()
)
kafka_df.printSchema()
```

```python
kafka_df.show()
```

```python
kafka_df = kafka_df.withColumns({
    "key":kafka_df.key.cast("string"),
    "value":kafka_df.value.cast("string")
})
```

```python
#kafka_df.show()
```

```python
from pyspark.sql.types import StructType , StructField , StringType , DoubleType , TimestampType
json_schema = StructType([
    StructField("BTCUSDT",  StringType(), nullable=True),
    StructField("ETHUSDT",  StringType(), nullable=True),
    StructField("PAXGUSDT", StringType(), nullable=True),
    StructField("SOLUSDT",  StringType(), nullable=True),
    StructField("datetime", TimestampType() , nullable=True)
])
```

```python
json_kafka = kafka_df.withColumn("json_value",from_json("value",json_schema)).select("json_value.*","timestamp")
```

```python
json_kafka.show()
```

```python
from uuid import uuid4
batch_df = json_kafka.withColumn("id", lit(str(uuid4()))).withColumnsRenamed({
    "datetime":"created_at",
    "timestamp":"kafka_timestamp",
    "BTCUSDT":"btcusdt",
    "ETHUSDT": "ethusdt",
    "PAXGUSDT":"paxgusdt",
    "SOLUSDT":"solusdt"
})
```

```python
#batch_df.show()
```

```python
def write_to_cassandra(batch_df,batch_id):
    if batch_df.isEmpty():
        return 

    try:
        (
            batch_df
            .write
            .format("org.apache.spark.sql.cassandra")
            .mode("append")
            .option("keyspace","binance")
            .option("table","prices")
            .save()
        )
        print(f"Batch Saved Succesuly with taile {batch_df.count()}")
    except Exception as e:
        print(f"ERROR - {str(e)}")
```

```python
(
    batch_df
    .writeStream
    .outputMode("append")
    .foreachBatch(write_to_cassandra)
    .option("truncate", False)
    .trigger(processingTime="5 seconds")
    .option("checkpointLocation", "checkpoint_dir_kafka")
    .start()
    .awaitTermination()
)
```
